```python
# Import required libraries
import tkinter as tk          # For GUI
import pyttsx3                # For text-to-speech
import datetime               # For current date and time
import webbrowser             # To open websites
import pywhatkit              # To play YouTube videos and other web-based actions

# -------------------------------
# Initialize the voice engine
# -------------------------------
engine = pyttsx3.init()  # This sets up the text-to-speech engine

# Function to convert text to speech
def speak(text):
    engine.say(text)          # Queue the text to be spoken
    engine.runAndWait()       # Speak the queued text

# ---------------------------------
# Main logic for the assistant
# ---------------------------------
def run_assistant():
    query = user_input.get().lower()  # Get user input and convert it to lowercase for easier matching

    # Check if the input contains the word "time"
    if "time" in query:
        time = datetime.datetime.now().strftime('%I:%M %p')  # Get current time in 12-hour format
        response = f"The time is {time}"                      # Form a response string

    # Check if the input contains the word "date"
    elif "date" in query:
        date = datetime.datetime.now().strftime('%B %d, %Y')  # Get current date
        response = f"Today's date is {date}"                  # Form a response string

    # Check if the input asks to open Google
    elif "open google" in query:
        webbrowser.open("https://www.google.com")             # Open Google in the default browser
        response = "Opening Google"

    # Check if the user wants to play something (e.g., a song)
    elif "play" in query:
        song = query.replace("play", "").strip()              # Extract the song name from the query
        pywhatkit.playonyt(song)                              # Play the song on YouTube
        response = f"Playing {song} on YouTube"

    # If input doesn't match any known command
    else:
        response = "Sorry, I didn't understand that."

    # Show the response in the GUI and also speak it
    display_result.set(response)     # Update the label in the GUI with the response
    speak(response)                  # Use text-to-speech to speak the response
    user_input.delete(0, tk.END)     # Clear the input box

# ---------------------------------
# GUI Setup using Tkinter
# ---------------------------------
app = tk.Tk()                        # Create the main application window
app.title("Smart Personal Assistant")  # Set window title
app.geometry("600x400")             # Set window size (width x height)
app.config(bg="#283747")            # Set background color

# StringVar to dynamically update GUI text
display_result = tk.StringVar()

# Title label
title = tk.Label(
    app,
    text="Smart Personal Assistant",
    font=("Arial", 20),
    bg="#283747",
    fg="white"
)
title.pack(pady=20)                 # Add some vertical space (padding) below the label

# Text input field for user queries
user_input = tk.Entry(
    app,
    font=("Arial", 14),
    width=40
)
user_input.pack(pady=10)

# "Ask" button to trigger the assistant
run_button = tk.Button(
    app,
    text="Ask",
    font=("Arial", 13),
    command=run_assistant,           # This button calls the run_assistant() function
    bg="#1ABC9C",
    fg="white"
)
run_button.pack(pady=10)

# Label to display assistant's response
output = tk.Label(
    app,
    textvariable=display_result,     # This will automatically update when display_result.set() is called
    font=("Arial", 16),
    bg="#283747",
    fg="#F7DC6F"
)
output.pack(pady=20)

# Exit button to close the application
exit_button = tk.Button(
    app,
    text="Exit",
    command=app.destroy,             # This will close the app
    bg="#E74C3C",
    fg="white",
    font=("Arial", 12)
)
exit_button.pack(pady=10)

# Start the Tkinter event loop (keep the window open)
app.mainloop()
```
