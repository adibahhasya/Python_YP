```python
# WAP to take 3 details (name, email, mobno) from 3 users using loops and lists

users = []  # list to store all users' details

# Taking details of 3 users
for i in range(3):
    print(f"\nEnter details for User {i+1}:")
    name = input("Name: ")
    email = input("Email: ")
    mobno = input("Mobile No: ")

    # store one user's info as a list inside the main list
    users.append([name, email, mobno])

# a) Print all information
print("\n--- All Users Information ---")
for i, user in enumerate(users, start=1):
    print(f"User{i}: Name = {user[0]}, Email = {user[1]}, Mobile No = {user[2]}")

# b) Print information for 2nd user
print("\n--- Information for 2nd User ---")
print(f"Name: {users[1][0]}")
print(f"Email: {users[1][1]}")
print(f"Mobile No: {users[1][2]}")

# c) Print mobile number of 3rd user
print("\n--- Mobile number of 3rd User ---")
print(f"Mobile No: {users[2][2]}")
```

using dictionary

```python
# Store user information
users = []

# Use loop to take input for 3 users
for i in range(1, 4):
    print(f"\nEnter details for User {i}:")
    name = input("Enter name: ")
    email = input("Enter email: ")
    mob_no = input("Enter mobile number: ")

    # Store the info as a dictionary
    user_info = {
        "Name": name,
        "Email": email,
        "Mobile No": mob_no
    }

    users.append(user_info)

# a) Print all information
print("\n--- All Users Information ---")
for i, user in enumerate(users, start=1):
    print(f"\nUser {i}:")
    for key, value in user.items():
        print(f"{key}: {value}")

# b) Print all information for 2nd user only
print("\n--- Information for 2nd User ---")
for key, value in users[1].items():
    print(f"{key}: {value}")

# c) Print all information for 3rd user only
print("\n--- Information for 3rd User ---")
for key, value in users[2].items():
    print(f"{key}: {value}")
```

```python
#YASh's solution

class takeUserDetails:

  def __init__(self,n,e,mn):     # __init__(c1,n,e,mn)           __init__(c2,n,e,mn)
    self.name = n                # c1.name = "John"               c2.name = "Alex"
    self.email = e               # c1.email = "j@gmail.com"       c2.email = "al@gmail.com"
    self.mobno = mn              # c1.mobno = 603564543           c2.mobno = 601424232

  def printDetails(self):                             # printDetails(c1)
    print(self.name,self.email,self.mobno)            # c1.name  c1.email  c1.mobno

c1 = takeUserDetails("John","j@gmail.com",603564543)
c2 = takeUserDetails("Alex","al@gmail.com",601424232)
c3 = takeUserDetails("Tina","tn@gmail.com",646362721)

# Printing all information
print('---- Information Of All Users ----')
c1.printDetails()
c2.printDetails()
c3.printDetails()

# Print the second user information
print('\n\n---- Information Of Second User----')
c2.printDetails()

# Print the mobno of 3rd user
print('\n\n---- Third User Mob No----')
print(c3.mobno)
```
