```python
# Exception handling in Python is a mechanism that allows you to handle runtime errors,
#  or exceptions, in a controlled way, rather than letting your program crash unexpectedly.
# Python provides a way to catch errors, respond to them, and allow your program to continue
# running smoothly even if something goes wrong.

```

```python
 #WITHOUT exception handling
# try except finally
# try : keep that block of code inside try , which might give you an error
# except : catch that error

# Without Exception Handling

val1 = int(input('Enter First Value : '))
val2 = int(input('Enter Second Value : '))
res = val1 / val2
print('After Division :',res)
print('Rest Of The Code')
```

```python
# Without Exception Handling

try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
  print('Rest Of The Code')

except:
  print('something went wrong')
print('rest of the code')
```

```python

try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
  print('Rest Of The Code')

except Exception as e:
  print('Error: ', e)
print('rest of the code')
```

```python
try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
  print('Rest Of The Code')

except ZeroDivisionError as e:
  print('Error: ', e)
print('rest of the code')
```

```python
# define multiple exceptions
# but only one exception will work at a same time

try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
except (ZeroDivisionError,ValueError) as e:
  print('Error :',e)
print('Rest Of The Code')
```

```python
#finally keyword

#it will execute always

try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
  print('Rest Of The Code')

except ZeroDivisionError as e:
  print('Error: ', e)
print('rest of the code')
```

```python
# Case 3 : when there is an exception but not handeled properly
try:
  val1 = int(input('Enter First Value : '))
  val2 = int(input('Enter Second Value : '))
  res = val1 / val2
  print('After Division :',res)
except ModuleNotFoundError as e:
  print('Error :',e)
finally:
  print('I will execute always')
print('Rest Of The Code')
```
