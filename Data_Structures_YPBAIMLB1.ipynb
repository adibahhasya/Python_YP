```python
#List Built-In Functions
#help(list)
mylist = [10, 20, 30, 40, 50, 60]

# a.) append(value)
# used to add a new element at the END of the List
print(mylist)
mylist.append(110)
print('After Append: ',mylist)

# b.) insert(indexNumber,value)
# used to add a new element at any Specific Position
mylist.insert(2,2000) #[index, no yang perlu ditukar]
print('After Insert: ',mylist)

# c.) pop(indexNumber)
#used to delete an element from the specified index number from the list
mylist.pop(2)
print('After Pop: ',mylist)


# d.) pop()
# if no argument is provided, it will delete the last element
mylist.pop()
print('After Pop: ',mylist)

# e.) remove(value)
mylist.remove(50)
print('After Remove: ',mylist)


#F) REVERSE()
mylist.reverse()    #PRINT(MYLIST[::-1])
print('After Reverse :', mylist)
print(mylist[::-1])

#g.) clear()
#it is used to delete all the element from list
mylist.clear()
print('After Clear :', mylist)

#h.) del keyword
del mylist
```

#homework, cuba replace 10 to 5

```python
#Method 1: Loop with index
numbers = [10,20,30,40,10,10,10,50,60,70,10,80,10]

for i in range(len(numbers)):
    if numbers[i] == 10:
        numbers[i] = 5   # replace 10 with 5

print(numbers)
# [5, 20, 30, 40, 5, 5, 5, 50, 60, 70, 5, 80, 5]
```

```python
#2 Method 2: List comprehension
numbers = [10,20,30,40,10,10,10,50,60,70,10,80,10]

numbers = [5 if x == 10 else x for x in numbers]

print(numbers)
# [5, 20, 30, 40, 5, 5, 5, 50, 60, 70, 5, 80, 5]
```

```python
#Method 3: Using enumerate() (more explicit)

numbers = [10,20,30,40,10,10,10,50,60,70,10,80,10]

for idx, val in enumerate(numbers):
    if val == 10:
        numbers[idx] = 5

print(numbers)
# [5, 20, 30, 40, 5, 5, 5, 50, 60, 70, 5, 80, 5]
```

```python
# NESTED LIST COMPREHENSION
#[[expression for item in inner_iterable] for item in outer_iterable]
# create a new list showing the square of mentioned numbers

numbers = [2,3,4,5,6,7,8,9,10]
sq_nums = []
for i in numbers:
  sq_nums.append(i**2)
print(sq_nums)
```

```python
#write like this instead of previous cell
numbers = [2,3,4,5,6,7,8,9,10]
sq_nums = [i**2 for i in numbers]
print(sq_nums)
```
