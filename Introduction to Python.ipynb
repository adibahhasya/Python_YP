```python
#sep used to separate value by using whitespace
print('python', 'programming', '')
print(78, 89, 101)
print('python', 'programming', sep=':)')


# sep used to separate value by using a whitespace
print('Python','Programming','Is','Fun','To','Learn')
print(78,89,101)
print('Python','Programming','Is','Fun','To','Learn',sep='-->')
```

```python
# end used to add \n [newline] at the end of the value
print('Python') #Python\n (\n means new line)
print('Programming')
print ('Nexperts', end='\t') #\t tabspace
print('Academy')
print('Yash')

#By default, python akan buat newline sendiri, if we need tabspace just use \t
```
